```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df = pd.read_csv('Aviation_traffic_data_with_Longtitude_and_Latitude.csv')
```

```python
num_cols = df.select_dtypes('number').columns.to_list()
num_cols
```

```python
numerical_cols = ['Departures_On_Time', 'Departures_Delayed', 'Dom_Pax_Total', 'Int_Pax_Total', 'Pax_Total']
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
df['Month_date'] = pd.to_datetime(df['Month_date'], format='%b-%y')
```

```python
# Plotting time series data
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month_date', y='Pax_Total', data=df, estimator='sum')
plt.title('Total Passenger Traffic Over Time')
plt.xlabel('Month')
plt.ylabel('Total Passengers')
plt.show()
```
